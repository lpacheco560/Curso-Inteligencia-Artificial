```python
import numpy as np

# Set random seed
np.random.seed(42)

# Create 'saldo' array with synthetic bank balance data
saldo = np.random.randint(1000, 10000, size=20)  # Random balances between 1000 and 10000

# Create 'antiguedad' array for bank account age
antiguedad = np.arange(0, 10, 0.5)  # Values from 0 to 10 with a step of 0.5

# Create 'ubicacion' matrix with latitude and longitude near Bucaramanga
latitud = 7.12 + np.random.rand(20) * 0.1  # Vary latitude around 7.12
longitud = -73.12 + np.random.rand(20) * 0.1  # Vary longitude around -73.12
ubicacion = np.column_stack((latitud, longitud))

# Combine using column_stack
datos_combinados = np.column_stack((saldo, antiguedad, ubicacion))

print(datos_combinados)
```
